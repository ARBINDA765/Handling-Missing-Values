# 2-Random Sample Imputation

Aim: 
1. Random sample imputation consists of taking random observation from the dataset and we use this observation to replace the nan values

2. Random Sample Imputation take a random observation from the feature.

3. After that we use random observation to replace NaN in that feature.



When should it be used? 


> It assumes that the data are missing completely at random(MCAR)

```python
import pandas as pd
df=pd.read_csv('titanic.csv', usecols=['Age','Fare','Survived'])
df.head()
```

```python
#Checking the null values 
df.isnull().sum()
```

```python
#% OF NULL values 
df.isnull().mean()*100
```

```python
#No of records are having null
df['Age'].isnull().sum()
```

```python
#DataFrame.sample(n=None, frac=None, replace=False, weights=None, random_state=None, axis=None, ignore_index=False)[source]Â¶
#Return a random sample of items from an axis of object.

df['Age'].sample(n=df['Age'].isnull().sum(),random_state=0)
```

```python

#- where .dropna() function drop all nan values in that feature 
#- This function check where is null values are present and replace Nan with random sample
# We use random_state because it replace NaN with specific value only ( if we not use random_state then values change evrytime when we run .
df['Age'].dropna().sample(df['Age'].isnull().sum(),random_state=0)
```

```python
df['Age'].isnull().head(6)
```

```python
#Finding the index where the value is null
df[df['Age'].isnull()].index
```

```python

#Create A function which Replace NaN with Random Sample
#Here we creating a two new features Age_median and Age_random,Age_Mean
#In Age_median replace NaN with Median and In Age_random replce NaN with Random Sample

def impute_nan(df,variable,median):
    df[variable+"_median"]=df[variable].fillna(median) #replace Nan with median,Fillna (will fill na value with the defined value)
    df[variable+'_mean']  = df[variable].fillna(df[variable].mean())
    df[variable+"_random"]=df[variable]##Copy feature into new feature
     
#calculate random sample and store into random_sample
##It will have the random sample to fill the nan value
#in random_sample all filled nan values are present now we want to put/merge this all filled values in our dataset
# for this we want to match all nan values index in random_sample with df[variable_'random] 
    
    random_sample=df[variable].dropna().sample(n=df[variable].isnull().sum(),random_state=0)
    ##pandas need to have same index in order to merge the dataset
    random_sample.index=df[df[variable].isnull()].index
    df.loc[df[variable].isnull(),variable+'_random']=random_sample
```

# 4- End of Distribution imputation

If there is suspicion that the missing value is not at random then capturing that information is important.
In this scenario, one would want to replace missing data with values that are at the tails of the distribution of the variable.

> In this method we replace missing values with far end values or extreme

> Far end value means the values after 3rd stander deviation

```python
df_m=pd.read_csv('titanic.csv', usecols=['Age','Fare','Survived'])
df_m.head()
```

```python
#Now we have a mean and stander deviation of Age feature now we find the Far end values or extreme value
#where extreme values means the values after 3rd stander deviation
#so extreme = (Mean of Age Feature + ( 3 * stander deviation of Age Feature)

extreme=df_m.Age.mean()+3*df_m.Age.std()
```

```python
df_m.Age.hist(bins=50)
```

```python
import seaborn as sns
sns.boxplot('Age',data=df_m)
```

# Observation:
    from both plots we can see that extrame values are between ~ 70 to 80 in other word it is a outliers

```python
def impute_nan(df_m,variable,extreme):
    df_m[variable+"_end_distribution"]=df_m[variable].fillna(extreme)
    df_m[variable+"_Median"]= df_m[variable].fillna(df_m[variable].median())#Age Nan value is replaced by Median 
```

```python
impute_nan(df_m,'Age',extreme)
```
